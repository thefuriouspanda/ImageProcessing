### As by default cv2 read images in BGR format so let's Convert from BGR to RGB

```python
cat1=cv2.cvtColor(cat,cv2.COLOR_BGR2RGB)
dog1=cv2.cvtColor(dog,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(cat1)
```

```python
plt.imshow(dog1)
```

### As both the images are of different resolutions so let's resize to same proportion 

```python
cat_new = cv2.resize(cat1,(500,500),cv2.INTER_LINEAR)
```

```python
dog_new = cv2.resize(dog1,(500,500),cv2.INTER_LINEAR)
```

### Let's display images

```python
plt.imshow(cat_new)
```

```python
plt.imshow(dog_new)
```

### Let's find the face areas of both the animals by hit and trial

```python
plt.imshow(cat_new[25:175,300:450])
```

```python
plt.imshow(dog_new[20:170,220:370])
```

### Let's replace the Cat face data with Dog face data

```python
dog_face = dog_new[20:170,220:370]
```

```python
 dog_new[20:170,220:370]
```

```python
cat_new[25:175,300:450] = dog_face
```

### Let's display the New Cat Image(Our Required Output)

```python
plt.imshow(cat_new)
```
